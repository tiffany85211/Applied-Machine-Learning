### Import libraries 

```python
%matplotlib inline
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
```

### Import Iris dataset

```python
data = pd.read_csv('./data/iris.data', header=None)
data.columns = ["Sepal-length", "Sepal-width", "Pedal-length", "Pedal-width", "Species"]
print(data.head())
print("Shape: " + str(data.shape))
```

### Visualization: Scatter Matrix Plot

#### Plotting using seaborn

```python
import seaborn as sns
sns.set()
sns.pairplot(data, hue="Species")
```

#### Plotting using Matplotlib

```python
# Create subplots
numvars = 4
fig, axes = plt.subplots(nrows=numvars, ncols=numvars, figsize=(8,8))
fig.subplots_adjust(hspace=0.1, wspace=0.1)

# Create label for coloring
color_dict = {'Iris-setosa': 'r', 'Iris-versicolor': 'g', 'Iris-virginica': 'b'}
labels = [ color_dict[i] for i in data.iloc[:,4] ]

# Plot scatter plots
for i, j in zip(*np.triu_indices_from(axes, k=1)):
    for x, y in [(i,j), (j,i)]:
        axes[y,x].scatter(data.iloc[:,x], data.iloc[:,y], c=labels, s=5)

# Deal with axes and formats
for ax in axes.flat:
    # Background color
    ax.set_facecolor('white')
    # Border lines
    ax.spines['top'].set_color('black')
    ax.spines['top'].set_linewidth(1)
    ax.spines['bottom'].set_color('black')
    ax.spines['bottom'].set_linewidth(1)
    ax.spines['left'].set_color('black')
    ax.spines['left'].set_linewidth(1)
    ax.spines['right'].set_color('black')
    ax.spines['right'].set_linewidth(1)

    # Hide all ticks and labels
    ax.xaxis.set_visible(False)
    ax.yaxis.set_visible(False)

    # Set up ticks only on one side for the "edge" subplots
    if ax.is_first_col():
        ax.yaxis.set_ticks_position('left')
    if ax.is_last_col():
        ax.yaxis.set_ticks_position('right')
    if ax.is_first_row():
        ax.xaxis.set_ticks_position('top')
    if ax.is_last_row():
        ax.xaxis.set_ticks_position('bottom')
        
# Label the diagonal subplots by Species Names
for i, label in enumerate(data.columns[:4]):
    axes[i,i].annotate(label, (0.5, 0.5), xycoords='axes fraction',
            ha='center', va='center')

# Set up ticks and labels for certain subplots
for i, j in [(0,1), (0,3), (3,0), (3,2)]:
    axes[i,j].xaxis.set_visible(True)
    axes[j,i].yaxis.set_visible(True)
    
fig.suptitle("Iris Data (red=setosa, green=versicolor, blue=virginica)")
```
